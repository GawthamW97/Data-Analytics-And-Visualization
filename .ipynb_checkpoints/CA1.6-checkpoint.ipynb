## Utilities

```python
def histogram(data, bins=None, xlim=None, title=None, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
        fig.set_size_inches(4,3)
    mean = stats.mean(data)
    sd = stats.pstdev(data)
    quantiles = stats.quantiles(data, n=1000)
    xmin = math.floor(quantiles[0])
    xmax = math.ceil(quantiles[-1])
    
    # Histogram
    if bins is None:
        bins = [x+0.5 for x in range(xmin,xmax,1)]

    if xlim is None:
        ax.set_xlim(xmin,xmax)
    else:
        ax.set_xlim(xlim[0], xlim[1])
        
    plt.hist(data, bins=bins, density=True)
    
    # pdf
    D = stats.NormalDist(mu=mean, sigma=sd)
    X = [ float(x) for x in np.linspace(xmin,xmax,1000) ]
    Y = [ D.pdf(x) for x in X ]
    plt.plot(X, Y, 'r-', lw=1)
        
    if title is not None:
        plt.title(title)
    plt.grid()
```

```python
import collections

def values(data):
    return list(set(data))

def labels(data):
    return [str(i) for i in values(data)]
    
def counts(data):
    c = collections.Counter(data)
    return [c[i] for i in values(data)]

def probabilities(data):
    c = collections.Counter(data)
    t = total(data)
    return [c[i]/t for i in values(data)]

def cumsum(l):
    if len(l)<2:
        return l
    cs = cumsum(l[:-1])
    return cs + [cs[-1]+l[-1]]
    
def total(data):
    return len(list(data))
```

```python
def frequencyPlot(c, title, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
        fig.set_size_inches(4,3)
    ax.bar(values(c), counts(c))
    ax.xaxis.set_ticks(values(c))
    ax.set_title(title)
    ax.grid(axis='y')
```

```python
def correlationPlot(data, b, a, regression=True):
    fig, ax = plt.subplots()
    fig.set_size_inches(4,3)
    
    a_mean = stats.mean(data[a])
    a_sd = stats.pstdev(data[a])
    a_quantiles = stats.quantiles(data[a], n=1000)
    xmin = math.floor(a_quantiles[0])
    xmax = math.ceil(a_quantiles[-1])
    ax.set_xlabel(a.title())
    
    b_mean = stats.mean(data[b])
    b_sd = stats.pstdev(data[b])
    b_quantiles = stats.quantiles(data[b], n=1000)
    ymin = math.floor(b_quantiles[0])
    ymax = math.ceil(b_quantiles[-1])
    ax.set_ylabel(b.title())

    X = [ float(x) for x in data[a] ]
    Y = [ float(y) for y in data[b] ]
    plt.plot(X, Y, 'bo', ms=1)

    plt.title(b.title()+" ~ "+a.title())

    if regression:
        results = sm.OLS(data[b], add_constant(data[a]), hasconst=True).fit()
        C = results.params['const']
        L = results.params[a]
        plt.plot([xmin, xmax], [C+L*xmin, C+L*xmax], 'r-')

    ax.grid()
    plt.show()
```

```python
def plot_Residual_Fitted(result, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
        fig.set_size_inches(4,4)
    
    ax.scatter(result.fittedvalues , result.resid, marker='.')
    ax.set_xlabel('Fitted value')
    ax.set_ylabel('Residual')
    ax.set_title('Residual vs Fitted')

    points = sorted(result.fittedvalues.unique())
    if len(points)<20:
        X = list(points)
        Y = [ result.resid[result.fittedvalues==x].mean() for x in X ]
        plt.plot(X, Y, 'r--', lw=0.5)
```

```python
def boxplots(df, y, by=None, w=4):
    fig, ax = plt.subplots()
    fig.set_size_inches(w,3)
    df.boxplot(y, by=by, ax=ax)
    ax.set_ylabel(y.title())
    if by is not None:
        ax.set_xlabel(by.title())
    ax.grid(axis='x')
    plt.title(y.title()+ (' by '+by.title() if by is not None else ''))
    plt.suptitle('')
    plt.show()
```

```python
from statsmodels.graphics.gofplots import qqplot

def plot_QQ(result, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
        fig.set_size_inches(4,4)

    plt.title('Q-Q Residuals')
    qqplot(result.resid, line='r', fit=True, ax=ax, marker='.')
    ax.set_ylabel('Standardized Residuals')
```

```python
def getAccuracy(X, y,t_size = 0.2 ,r_state = 42, m_iter=1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = t_size, random_state=r_state)
    logreg = LogisticRegression(max_iter = m_iter)
    logreg.fit(X_train, y_train)
    y_pred = logreg.predict(X_test)
    print('Accuracy of logistic regression classifier on test set: {:2f}'.format(logreg.score(X_test,y_test)))
    return y_pred
```

```python
#From the lectrue notes
from statsmodels.stats.outliers_influence import variance_inflation_factor

def VIF(X):
    VIF = pd.DataFrame()
    for i in range(X.shape[1]):
        VIF[X.columns[i]] = [ round(float(variance_inflation_factor(X, i)), 1) ]
    return VIF
```

```python
#range encode columns with high cardinality
# https://stackoverflow.com/questions/44151071/how-to-encode-a-range-of-values-using-pandas
def rangeEncode(df,x):
    maximum = df[x].max()
    bins = [0, round(maximum * 1/4), round(maximum * 2/4), round(maximum * 3/4), maximum + 1]
    # labels=[1,2,3,4]
    labels=["0 - "+str(round(maximum * 1/4)),str(round(maximum * 1/4)) +"-"+str(round(maximum * 2/4)),str(round(maximum * 2/4)) +"-"+str(round(maximum * 3/4)),str(round(maximum * 3/4)) +"-"+str(maximum + 1)]
    
    df[x] = pd.cut(df[x], bins=bins, labels=labels, include_lowest=True)
```

```python
def zScoreTransformation(df,column_lst):
    columnNames = []
    for x in column_lst:
        columnName = 'z_'+x
        df[columnName] = zscore(df[x])
        threshold = 3
        df = df[abs(df[columnName]) <= threshold]
        columnNames.append(columnName)
    df = df.drop(columnNames,axis=1)
    return df
```

```python
def labelCategoricalColumns(df,lst):
    for x in lst:
        # for all the columns which is of data type Category we Label Encoded them
        if(x in medication_columns):
            dosage_mapping = {}
            if("Up" in df[x].values):
                dosage_mapping["Up"] = 2
            if("Down" in df[x].values):
                dosage_mapping["Down"] = -1
            if("Steady" in df[x].values):
                dosage_mapping["Steady"] = 1
            if(' No' in df[x].values):
                dosage_mapping[" No"] = 0
            od[x] =  od[x].map(dosage_mapping)
        else:
            df[x] = df[x].astype('category').cat.codes
    return df
lebeled_df = labelCategoricalColumns(od,categorical_columns)
```
