```python
check_query = """
OPTIONAL MATCH (n) WITH n LIMIT 1 
RETURN n IS NOT NULL AS IsNodesPresent
"""
country_query = """
MATCH (n:Country) RETURN n.code as Code, n.zone as Zone, n.name as Name LIMIT 25
"""

emission_by_country_query = """
MATCH (c:Country)-[r:HAS_PRODUCED]->(e:Emission)
WITH c,e,r
RETURN
    c.code as Code,
    c.zone as Zone,
    c.name as Name,
    e.ipcc_code as IPCC_Code,
    e.ipcc_name as IPCC_Code_Name LIMIT 25
"""

emission_by_year = """
MATCH (c:Country)-[r:HAS_PRODUCED]->(e:Emission)
WITH c,e,r
WHERE c.code = {$value}
RETURN
    c.code as Code,
    c.name as Name 
    properties(e) LIMIT 5
"""
```

```python
def connectToDataBases():
    URI = "bolt://localhost:7687"
    AUTH = ("neo4j", "asd123asd123")
    
    driver = GraphDatabase.driver(URI, auth=AUTH)
    return driver
```

```python
def getCountry(query,name=""):
    with driver.session() as session: 
        result = session.run(query)
        data = [dict(record) for record in result]
        return pd.DataFrame(data)
```

```python
def getRecords(query,value=""):
    with driver.session() as session: 
        if(len(value) > 0):
            result = session.run(query)
        else:
            result = session.run(query)
        data = [dict(record) for record in result]
        return pd.DataFrame(data)
```
