```python
from tqdm import tqdm
def insert_energy_stats_bulk(df, cursor, conn, columns=None, batch_size=500):

    # Default energy columns if not specified
    if columns is None:
        columns = ['iso_code','year','population','gdp','biofuel_cons_per_capita','biofuel_consumption','biofuel_electricity','biofuel_share_elec','carbon_intensity_elec','coal_cons_per_capita','coal_consumption','coal_electricity','coal_production','coal_share_elec','electricity_demand',
'electricity_generation','fossil_electricity','fossil_energy_per_capita','fossil_fuel_consumption','fossil_share_elec','gas_consumption','gas_energy_per_capita','gas_production',
'gas_share_elec','greenhouse_gas_emissions','low_carbon_consumption','low_carbon_electricity','low_carbon_energy_per_capita','low_carbon_share_elec','nuclear_consumption','nuclear_electricity','nuclear_energy_per_capita',
'nuclear_share_elec','oil_consumption','oil_electricity','oil_energy_per_capita','oil_production','oil_share_elec','other_renewable_consumption','other_renewable_electricity','other_renewables_energy_per_capita',
'other_renewables_share_elec','renewables_consumption','renewables_electricity','renewables_energy_per_capita','renewables_share_elec']
    
    # Keep only relevant columns
    energy_df = df[columns].copy()
    
    # Drop rows without iso_code or year (essential keys)
    energy_df = energy_df.dropna(subset=['iso_code', 'year'])
    energy_df = energy_df.replace({np.nan: None})
    
    print("Inserting energy stats into database...")

    placeholders = ', '.join(['%s'] * len(columns))
    col_names = ', '.join(columns)
    
    count = 0

    for _, row in tqdm(energy_df.iterrows(), total=len(energy_df)):
        try:
            values = [row[col] for col in columns]
            cursor.execute(f"""
                INSERT INTO energy_stats ({col_names})
                VALUES ({placeholders})
            """, values)
            count += 1

            if count % batch_size == 0:
                conn.commit()

        except Exception as e:
            print(f"Error inserting row: {e}")

    conn.commit()
    print(f"Inserted successfully. Total rows inserted: {count}")
```
